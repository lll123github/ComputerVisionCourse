```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class Layer:
    def __init__(self, w, b):
        self.w = w
        '''与上一层连接的权重'''
        self.b = b
        '''与上一层连接的偏置'''
        self.a = None
        '''上一层的输出'''
        self.y = None
        '''这一层的输出'''
        self.z = None
        '''这一层的输入'''
        self.dz = None
        self.dw = None
        self.db = None
        self.da = None

    def forward_prop(self, a):
        self.a = a
        self.z = np.dot(self.w, self.a) + self.b
        self.y = sigmoid(self.z)
        return self.y

    def back_prop(self, dy):
        self.dy = dy
        self.dz = self.dy * self.y * (1 - self.y)
        self.dw = np.dot(self.dz, self.a.T)
        self.db = self.dz
        self.da = np.dot(self.w.T, self.dz)
        return self.da


class Network:

    def __init__(self):
        self.layer1=Layer(w=np.ones((2,2))+np.random.rand(2,2),b=np.random.rand(2,1))
        self.layer2=Layer(w=np.ones((1,2))+np.random.rand(1,2),b=np.random.rand(1,1))
        # self.layer1=Layer(w=np.ones((2,2)),b=np.zeros((2,1)))
        # self.layer2=Layer(w=np.ones((1,2)),b=np.zeros((1,1)))
        self.layer_list=[self.layer1,self.layer2]
        self.batch_size=0
        self.y_array=np.ndarray((1,0))
        self.y_predict_array=np.ndarray((1,0))

    def forward_prop(self,a):
        for layer in self.layer_list:
            a=layer.forward_prop(a)
        return a
    def back_prop(self,da):
        for layer in reversed(self.layer_list):
            da=layer.back_prop(da)
        return da
    def train(self,x,y,index,learning_rate=0.01,batch_size=4):
        if not self.batch_size:
            self.batch_size=batch_size
            self.y_array=np.ndarray((1,self.batch_size))
            self.y_predict_array=np.ndarray((1,self.batch_size))
        y_predict=self.forward_prop(x)
        self.y_array[0,index%self.batch_size]=y
        self.y_predict_array[0,index%self.batch_size]=y_predict
        if(index%10000==0):
            print(f'第{index}次训练')
            e=np.sum((self.y_predict_array-self.y_array)**2)/2
            print(f'损失为{e}')
            print(f'真实值为{self.y_array}')
            print(f'预测值为{self.y_predict_array}')
        # if(index%self.batch_size==0):
        self.back_prop(y_predict-y)
        for layer in self.layer_list:
            layer.w-=learning_rate*layer.dw
            layer.b-=learning_rate*layer.db
    def outputLayerParameter(self):
        print('参数：')
        for layer in self.layer_list:
            print(layer.w,layer.b)
    def predict(self,x):    
        return self.forward_prop(x)




network=Network()
x_input=np.array([[0,1],[1,0],[0,0],[1,1]]).T
y_input=np.array([1,1,0,0])
for i in range(1000000):
    network.train(x_input[:,i%4].reshape(-1,1),y_input[i%4],i,learning_rate=1)

network.outputLayerParameter()
for i in range(4):
    print(f'输入为{x_input[:,i]}，输出为{network.predict(x_input[:,i].reshape(-1,1))}')
```
